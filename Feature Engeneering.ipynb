```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('cirrhosis.csv')
df.head(2)
```

Feature Engeneering

```python
df.info()
```

```python
df.describe()
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
##Replacing missing values of categorical features with their respective mode 
df.Drug.fillna(df.Drug.mode()[0],inplace=True)
df.Ascites.fillna(df.Ascites.mode()[0],inplace=True)
df.Hepatomegaly.fillna(df.Hepatomegaly.mode()[0],inplace=True)
df.Spiders.fillna(df.Spiders.mode()[0],inplace=True)
df.Stage.fillna(df.Stage.mode()[0],inplace=True)

##Replacing missing values of numerical features with their respective mean
df.Tryglicerides.fillna(df.Tryglicerides.mean(),inplace=True)
df.Platelets.fillna(df.Platelets.mean(),inplace=True)
df.Prothrombin.fillna(df.Prothrombin.mean(),inplace=True)
df.SGOT.fillna(df.SGOT.mean(),inplace=True)
df.Alk_Phos.fillna(df.Alk_Phos.mean(),inplace=True)
df.Copper.fillna(df.Copper.mean(),inplace=True)
df.Cholesterol.fillna(df.Cholesterol.mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Status'].value_counts()
#No severe imbalance Present
```

```python
df.drop('ID',axis=1,inplace=True) #Removing irrelevant feature
```
